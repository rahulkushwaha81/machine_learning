In this notebook, we will analyse how by increasing the polynomial features of a dataset helps in creating a model with better prediction capability.

We will do this analysis using linear regression.

## Generate random data set with polynomial flavour

```python
#importing all the modules at one place
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error

#generate random value of x
x = np.random.randint(-100,100,1000).reshape(1000,1)
#derive non-linear relationship between x and y
y = ((0.5)*x**3-x**2) - np.random.randint(1,10000) 
```

```python
#let us pack above data into a dataframe for the analysis
df = pd.DataFrame(x,columns=['x'])
df['y'] = y
df.head(5)
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(12,4))
sns.lineplot(x=df['x'],y=df['y'],data=df,ax=ax[0])
ax[0].set(title='Non-Linear relationship between x and y')
# if we try to derive a linear relationship through seaborn regplot, this how it will look.
# Check the straight line in the second graph.This is the best line fit.
sns.regplot(x=df['x'],y=df['y'],data=df,ax=ax[1])
ax[1].set(title='Calculated Linear relationship between x and y')
```

## Let's create a linear model (degree=1)

```python
#it's always good to reshape the data into numpy arrays
X = df['x'].values.reshape(1000,1) 
y = df['y'].values.reshape(1000,1)

#split data into training and test sets
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3) #splitting
```

```python
#train the model on training set
model = LinearRegression() #create the instance of linear model
model.fit(X_train,y_train) #fit/train the model on training set
```

```python
#predict the data on test set and not on the training set as model has already seen the training set.
model_prediction = model.predict(X_test)
```

**visualize the model prediction verses actual output**

```python
#packing model prediction and actual ouput into a dataframe for analysis and visualization
df_output = pd.DataFrame(X_test,columns=['X_test'])
df_output['y_test'] = y_test
df_output['model_prediction'] = model_prediction
df_output.head(5)
```

Predicted output is no where close to the real value.

Let's plot them

```python
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(12,4))
sns.lineplot(data=df_output,x='y_test',y='y_test',ax=ax[0])
ax[0].set(title='Ideal plot')
sns.lineplot(data=df_output,x='model_prediction',y='y_test',ax=ax[1])
ax[1].set( title='Derived plot')
plt.tight_layout()
```

Ideal plot should be the **straight line** between predicted output and actual output.

```python
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(12,4))
sns.lineplot(data=df_output,x='X_test',y='y_test',ax=ax[0])
ax[0].set(title='Actual plot between x and y for test set')
sns.lineplot(data=df_output,x='X_test',y='model_prediction',ax=ax[1])
ax[1].set( title='Derived Linear relationship plot')
plt.tight_layout()
```

Derived plot is not at all the best fit line as it is higly baised and we need to add more features to input data.

**Evaluating the model by calculating RMSE**

```python
print(np.sqrt(mean_squared_error(y_test,model_prediction))) # this is RMSE (root mean squared error)
print(y_test.mean())
```

As we can see that we can't compare RMSE against mean of output as the values of y_test varies tremendously between high positive and high negative values. Also RMSE looks very high for this model.

## Let's create a linear model with degree= 2

```python
poly_2 = PolynomialFeatures(degree=2) #create the instance 
poly_2.fit(X_train) #train it on the training set
X_train_poly_2 = poly_2.transform(X_train) #transform both training and test set
X_test_poly_2 = poly_2.transform(X_test)
X_train_poly_2[0:2] #have a look at transformed sample training set. 2 extra features have been added.
```

```python
model_poly_2 = LinearRegression()
model_poly_2.fit(X_train_poly_2,y_train)
model_poly_2_pred = model_poly_2.predict(X_test_poly_2)
```

```python
print(np.sqrt(mean_squared_error(y_test,model_poly_2_pred)))
print(y_test.mean())
```

There is little reduction in RMSE but not as much as we need.

```python
df_output['model_poly_2_pred'] = model_poly_2_pred
df_output.head(5)
```

For degree=2, model prediction is no where close to the actual output.

```python
fig,ax = plt.subplots(nrows=1,ncols=3,figsize=(12,4))
sns.lineplot(data=df_output,x='y_test',y='y_test',ax=ax[0])
ax[0].set(title='Ideal plot')
sns.lineplot(data=df_output,x='model_prediction',y='y_test',ax=ax[1])
ax[1].set( title='Actual plot degree=1')
sns.lineplot(data=df_output,x='model_poly_2_pred',y='y_test',ax=ax[2])
ax[2].set( title='Actual plot degree=2')
plt.tight_layout()
```

Both the above models are not close to best fit line as we are expecting a straight line between actual and predicted value.

**Let us add polynomial features of degree=3**

```python
poly_3 = PolynomialFeatures(degree=3)
poly_3.fit(X_train)
X_train_poly_3 = poly_3.transform(X_train)
X_test_poly_3 = poly_3.transform(X_test)
X_train_poly_3[0:2]
```

```python
model_poly_3 = LinearRegression()
model_poly_3.fit(X_train_poly_3,y_train)
model_poly_3_pred = model_poly_3.predict(X_test_poly_3)
```

```python
print(np.sqrt(mean_squared_error(y_test,model_poly_3_pred)))
print(y_test.mean())
```

**We can see that there is tremendous decrease in RMSE**

```python
df_output['model_poly_3_pred'] = model_poly_3_pred
df_output.head(5)
```

For degree=3, model prediction is extacly the same actual output.

```python
#plotting between model prediction vs actual ouput
fig,ax = plt.subplots(nrows=1,ncols=4,figsize=(12,4))
sns.lineplot(data=df_output,x='y_test',y='y_test',ax=ax[0])
ax[0].set(title='Ideal plot')

sns.lineplot(data=df_output,x='model_prediction',y='y_test',ax=ax[1])
ax[1].set( title='Derived plot degree=1')

sns.lineplot(data=df_output,x='model_poly_2_pred',y='y_test',ax=ax[2])
ax[2].set( title='Derived plot degree=2')

sns.lineplot(data=df_output,x='model_poly_3_pred',y='y_test',ax=ax[3])
ax[3].set( title='Derived plot degree=3')
plt.tight_layout()
```

We can see that plot for degree =3 is almost a straight line. This is what we are looking for.

```python
#plotting between test data and model prediction with different degrees
fig,ax = plt.subplots(nrows=1,ncols=4,figsize=(12,4))
sns.lineplot(data=df_output,x='X_test',y='y_test',ax=ax[0])
ax[0].set(title='Actual plot of test data')

sns.lineplot(data=df_output,x='X_test',y='model_prediction',ax=ax[1])
ax[1].set( title='Derived plot degree=1')

sns.lineplot(data=df_output,x='X_test',y='model_poly_2_pred',ax=ax[2])
ax[2].set( title='Derived plot degree=2')

sns.lineplot(data=df_output,x='X_test',y='model_poly_3_pred',ax=ax[3])
ax[3].set( title='Derived plot degree=3')
plt.tight_layout()
```

For degree=1 and 2, model is baised. 
For degree=3 plot looks same as the actual plot

## **Let us calculate RMSE on different polynomial degrees**

```python
#calculate RMSE for both training and test set to analyse the model
rmse_index = []
rmse_value_testset = [] #RMSE for test set
rmse_value_trainingset = [] #RMSE for training set
for i in range(0,21):
    poly_n = PolynomialFeatures(degree=i)
    poly_n.fit(X_train)
    X_train_poly_n = poly_n.transform(X_train)
    X_test_poly_n = poly_n.transform(X_test)
    model_poly_n = LinearRegression()
    model_poly_n.fit(X_train_poly_n,y_train)
    model_poly_n_pred_testset = model_poly_n.predict(X_test_poly_n)
    model_poly_n_pred_trainset = model_poly_n.predict(X_train_poly_n)
    rmse_index.append(i)
    rmse_value_testset.append(np.sqrt(mean_squared_error(y_test,model_poly_n_pred_testset)))
    rmse_value_trainingset.append(np.sqrt(mean_squared_error(y_train,model_poly_n_pred_trainset)))
```

```python
df_rmse = pd.DataFrame(rmse_index,columns=['degree'])
df_rmse['RMSE_test_set'] = rmse_value_testset
df_rmse['RMSE_training_set'] = rmse_value_trainingset
df_rmse.head(5)
```

```python
axp = df_rmse[['RMSE_test_set','RMSE_training_set']].plot()
axp.set(xlabel='degree',ylabel='RMSE')
```

## Conclusion:

RMSE for both training and test set starts decreasing as the degree reaches untill 2. It remains almost constant till degree = 10 and then starts increasing.

**Bias/Underfitting**
We can see that model is *baised* or undefitting for degree 1 and 2.

**Overfitting**
Model starts *overfitting* from degree =10 onwards as there is increase in error for both test and training set.

Ideal degree for the model should be 3.
